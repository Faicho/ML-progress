```python
def InsertSort(array):
  for step in range(1, len(array)):
    key = array[step]
    j  = step - 1
    while j >=0 and key < array[j]:
      array[j + 1] = array[j]
      j = j - 1

      array[j+1] = key
      print(array)

data = [1,2,31,3,4,1,3]
InsertSort(data)
```

```python
class Queue:
  def __init__(self):
    self.items = []

  def isEmpty(self):
    return self.items == []

  def enqueue(self, item):
    self.items.insert(0, item)

  def dequeue(self):
    return self.items.pop()

  def size(self):
    return len(self.items)
```
