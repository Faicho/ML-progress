```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(10,5))
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(df. Year, df. USA, 'r^-', label='United States of America')
plt.plot(df. Year, df. Canada, 'g.-', label='Canada')
plt.plot(df. Year, df['South Korea'], 'y.-',label='South Korea')
plt.plot(df. Year, df. Australia, 'b.-', label='Australia')

plt.xticks(df.Year[::1].tolist())

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend (loc = 'upper left')

plt.savefig('Gas_price_figure.png', dpi=500)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
fifa = pd.read_csv('fifa_data.csv')
plt.figure(figsize=(8,5), dpi=100)
plt.style.use('default')
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
print(fifa.Weight)

light =  fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125(light)', '125-150(light_medium)', '150-175(medium)', '175-200 (medium_heavy)', 'over 200 (heavy)']
explode = (0.04,0.04,0,0,0.04)
plt.title('Weight of Professional Soccer Players (lbs)')
plt.pie(weights, labels=label, explode=explode, radius = 1.8,pctdistance=0.8, autopct='%.2f %%')

#section between slices
circlebt = plt.Circle((0,0),0.7,color='white')
pl = plt.gcf()
pl.gca().add_artist(circlebt)

#plt.savefig("weights.png", dpi = 3000)
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
print(left)
print(right)
plt.figure(figsize=(8,5))
labels = ['Left', 'Right']
colors = ['pink', '#aabbcc']
plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f%%')
plt.title('Foot Preference of FIFA Players')
plt.show()
```
