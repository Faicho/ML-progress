```python
def BinarySearch(data, x, low, high):
    if high >= low:
        mid = low+(high-low)//2

        if data[mid] == x:
            return mid
        elif data[mid] > x:
            return BinarySearch(data, x, low, mid-1)
        else:
            return BinarySearch(data, x, mid + 1, high)
    else:
        return -1

arr = [12,17,18,34,67,89,90]
x = 12
y = 34
low = 0
high = len(arr) - 1
result = BinarySearch(arr, x, low, high)
result1 = BinarySearch(arr, y, low, high)
if result != -1 and result1 != -1:
    print("Number found at index", result)
    print("Number found at index", result1)
else:
    print("Not found")
```
